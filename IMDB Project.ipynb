```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer, make_column_selector
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error


import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
basics_original = pd.read_csv('https://datasets.imdbws.com/title.basics.tsv.gz', sep='\t', low_memory=False)
basics = basics_original.copy()
basics.head()
```

```python
akas_original = pd.read_csv('https://datasets.imdbws.com/title.akas.tsv.gz', sep='\t', low_memory=False)
akas = akas_original.copy()
akas.head()
```

```python
ratings_original = pd.read_csv('https://datasets.imdbws.com/title.ratings.tsv.gz', sep='\t', low_memory=False)
ratings = ratings_original.copy()
ratings.head()
```

```python
basics_original.to_csv('basics.csv.gz', compression='gzip',index=False)
akas_original.to_csv('akas.csv.gz', compression='gzip', index=False)
ratings_original.to_csv('ratings.csv.gz', compression='gzip', index=False)
```

### Preprocess Data

```python
# replace \n with nan
basics.replace({'\\N':np.nan}, inplace=True)
akas.replace({'\\N':np.nan}, inplace=True)
ratings.replace({'\\N':np.nan}, inplace=True)
```

```python
# drop movies missing run time and genre
basics.dropna(subset=['runtimeMinutes','genres','startYear'], axis=0, inplace=True)
```

```python
# filter out non-movies

movie_filter = basics['titleType'] == 'movie'
basics = basics.loc[movie_filter, :]
```

```python
basics.head()
```

```python
basics['startYear'] = basics['startYear'].astype('int64')
```

```python
# keep movies made between 2000 and 2022
filter_2000 = basics['startYear'] > 1999
filter_2022 = basics['startYear'] < 2023

basics = basics[filter_2000 & filter_2022]
```

```python
# Exclude movies that are included in the documentary category.
is_documentary = basics['genres'].str.contains('documentary',case=False)
basics = basics[~is_documentary]
```

```python
basics.head()
```

```python
filter_US = akas['region'] == 'US'
akas = akas[filter_US]
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
keepers_basics = basics['tconst'].isin(akas['titleId'])
```

```python
basics = basics[keepers_basics]
basics.head()
```

```python
ratings.head()
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
keepers_ratings = ratings['tconst'].isin(akas['titleId'])
```

```python
ratings = ratings[keepers_ratings]
ratings.head()
```
