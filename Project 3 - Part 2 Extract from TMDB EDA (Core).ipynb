# Exploratory Data Analysis

## Import Libraries | Download Data


```python
import pandas as pd

#import os, time,json
#from tqdm.notebook import tqdm_notebook
#import tmdbsimple as tmdb
```

```python
# Load in the dataframe from project 2 part 3:
df_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
```

```python
frames = [df_2000, df_2001]
df_combined = pd.concat(frames)
```

```python
df_combined
```

```python
df_combined.drop_duplicates(subset=['imdb_id'], inplace=True)
```

```python
df_combined.dropna(how='all', inplace=True)
```

```python
df_combined['imdb_id'].value_counts().sum()
```

```python
df_combined.info()
```

## Number of mofies with either revenue or budget data

```python
budget_filter = df_combined['budget'] > 0
revenue_filter = df_combined['revenue'] > 0
```

```python
df_financial_data = df_combined[budget_filter | revenue_filter]
```

```python
df_financial_data.value_counts('imdb_id').sum()
```

Number of movies with either revenue or budget data or both: 628

## Number of movies of each rating

```python
df_financial_data_id = df_financial_data[['certification', 'imdb_id']]
```

```python
df_financial_data_id.groupby(['certification']).count()
```

## Average revenue per rating

```python
df_financial_data_revenue = df_financial_data[['certification', 'revenue']]
```

```python
pd.options.display.float_format = '{:20,.0f}'.format
df_financial_data_revenue.groupby(['certification']).mean()
```

## Average budget per rating

```python
df_financial_data_budget = df_financial_data[['certification', 'budget']]
```

```python
df_financial_data_budget.groupby(['certification']).mean()
```
